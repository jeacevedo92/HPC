# <span style="color:green">Tile (Thread Block) Size Considerations</span>
<h2>
<div style="text-align:justify">
<ul>
<li>
Each <span style="color:orange">thread block</span> should have many threads
<ul>
<li>TILE_WIDTH of 16 gives 16*16 = 256 threads</li>
<li>TILE_WIDTH of 32 gives 32*32 = 1024 threads</li>
</ul>
</li>
<li>
    For 16, in each phase, each block performs 2*256 = 512 float loads from global memory for 256 * (2*16) = 8,192 mul/add operations. (16 floating-point operations for each memory load)
</li>
<li>
For 32, in each phase, each block performs 2*1024 = 2048 float loads from global memory for 1024 * (2*32) = 65,536 mul/add operations. (32 floating-point operation for each memory load)
</li>
</ul>
</h2>
</div>
<hr style="height:2px">

# <span style="color:green">Shared Memory and Threading</span>
<h2>
<div style="text-align:justify">
<ul>
<li>
Shared Memory and Threading
<ul>
<li>Shared memory size is implementation dependent!</li>
<li>TFor TILE_WIDTH = 16, each thread block uses 2*256*4B = 2KB of shared memory.</li>
<li>For 16KB shared memory, one can potentially have up to 8 thread blocks
executing
    <ul>
        <li>This allows up to 8*512 = 4,096 pending loads. (2 per thread, 256 threads per block)</li>
    </ul>
</li>
<li>The next TILE_WIDTH 32 would lead to 2*32*32*4 Byte= 8K Byte shared memory usage per thread block, allowing 2 thread blocks active at the same time
    <ul>
        <li>However, the thread count limitation of 1536 threads per SM in current generation GPUs will reduce the number of blocks per SM to one!</li>
    </ul>
</li>

</ul>
</li>
<li>Each ```__syncthread()``` can reduce the number of active threads for a block
    <ul>
        <li>More thread blocks can be advantageous</li>
    </ul>
</li>
</ul>
</h2>
</div>
