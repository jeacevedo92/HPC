# <span style="color:green">Partition and Summarize</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <ul>
            <body>
            
            <li>A commonly used strategy for processing large input
data sets
                <ul>
                    <li>There is no required order of processing elements in a data set (associative and
commutative)</li>
                    <li>Partition the data set into smaller chunks</li>
                    <li>Have each thread to process a chunk</li>
                    <li>Use a reduction tree to summarize the results from each chunk into the final
answer</li>
                </ul>
            </li>
            <li>Google and Hadoop MapReduce frameworks support
this strategy</li>
            <li>We will focus on the reduction tree step for now</li>
            </body>
        </ul>
</div>

<hr style="height:2px">

# <span style="color:green">Reduction enables other techniques</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <ul>
            <body>
            <li>Reduction is also needed to clean up after some
commonly used parallelizing transformations</li>
            <li>Privatization
                <ul>
                    <li>Multiple threads write into an output location</li>
                    <li>Replicate the output location so that each thread has a private output
location</li>
                    <li>Use a reduction tree to combine the values of private locations into
the original output location</li>
                </ul>
            </li>
            </body>
        </ul>
</div>

# <span style="color:green">What is a reduction computation?</span>

<br></br>
<div style="text-align:justify;font-size:20px">
       <body>
       <ul>
            
            <li>Summarize a set of input values into one value using a
“reduction operation”
                <ul>
                    <li>Max</li>
                    <li>Min</li>
                    <li>Sum</li>
                    <li>Product</li>
                </ul>
            </li>
            <li>Often used with a user defined reduction operation
function as long as the operation
                <ul>
                    <li>Is associative and commutative</li>
                    <li>Has a well-defined identity value (e.g., 0 for sum)</li>
                    <li>For example, the user may supply a custom “max” function for 3D
coordinate data sets where the magnitude for the each coordinate
data tuple is the distance from the origin.</li>
                </ul>
            </li>
            <br></br>
            
        </ul>
        <p style="font-size:30px"><strong>An example of “collective operation”</strong></p>
        </body>
</div>

# <span style="color:green">An Efficient Sequential Reduction O(N)</span>

<br></br>
<div style="text-align:justify;font-size:20px">
       <body>
       <ul>
            
            <li>Initialize the result as an identity value for the reduction
operation
                <ul>
                    <li>Smallest possible value for max reduction</li>
                    <li>Largest possible value for min reduction</li>
                    <li>0 for sum reduction</li>
                    <li>1 for product reduction</li>
                </ul>
            </li>
            <li>Iterate through the input and perform the reduction operation
between the result value and the current input value
                <ul>
                    <li>N reduction operations performed for N input values</li>
                    <li>Each input value is only visited once – an O(N) algorithm</li>
                    <li>This is a computationally efficient algorithm.</li>
                </ul>
            </li>
            
        </ul>
        </body>
</div>

# <span style="color:green">A Quick Analysis</span>

<br></br>
<div style="text-align:justify;font-size:20px">
       <body>
       <ul>
            
            <li>For N input values, the reduction tree performs
                <ul>
                    <li>(1/2)N + (1/4)N + (1/8)N + ... (1)N = (1- (1/N))N = N-1 operations</li>
                    <li>In Log (N) steps – 1,000,000 input values take 20 steps
                        <ul>
                            <li>Assuming that we have enough execution resources</li>
                        </ul>
                    </li>
                    <li>Average Parallelism (N-1)/Log(N))
                        <ul>
                            <li>For N = 1,000,000, average parallelism is 50,000</li>
                            <li>However, peak resource requirement is 500,000</li>
                            <li>This is not resource efficient</li>
                        </ul>
                    </li>
                </ul>
            </li>
            <li>This is a work-efficient parallel algorithm
                <ul>
                    <li>The amount of work done is comparable to the an efficient sequential algorithm</li>
                    <li>Many parallel algorithms are not work efficient</li>
                </ul>
            </li>
            
        </ul>
        </body>
</div>
