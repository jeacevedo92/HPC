# <span style="color:green">(Original) PCI Bus Specification</span>


<br></br>
<div style="text-align:justify;font-size:23px">

        
        
        <ul>
            <li>Connected to the Southbridge
                   <ul>
                       <li>Originally 33 MHz, 32-bit wide, 132 MB/second peak transfer rate</li>
                       <li>More recently 66 MHz, 64-bit, 528 MB/second peak</li>
                        <li>Upstream bandwidth remain slow for device (~256 MB/s peak)</li>
                        <li>Shared bus with arbitration
                            <ul>
                                <li>Winner of arbitration becomes bus master and can connect to CPU or DRAM
through the Southbridge and Northbridge</li>
                            </ul>
                        </li>
                   </ul>
            </li>

# <span style="color:green">PCIe 2 Links and Lanes</span>


<br></br>
<div style="text-align:justify;font-size:23px">

        
        
        <ul>
            <li>Each link consists of one or
more lanes
                   <ul>
                       <li>Each lane is 1-bit wide (4 wires,
each 2-wire pair can transmit
2.5Gb/s in one direction)
                            <ul>
                                <li>Upstream and downstream now
simultaneous and symmetric</li>
                            </ul>
                        </li>
                       <li>Each Link can combine 1, 2, 4,
8, 12, 16 lanes- x1, x2, etc.</li>
                        <li>Each byte data is 8b/10b
encoded into 10 bits with equal
number of 1’s and 0’s; net data
rate 2 Gb/s per lane each way</li>
                        <li>Thus, the net data rates are 250
MB/s (x1) 500 MB/s (x2), 1GB/s
(x4), 2 GB/s (x8), 4 GB/s (x16),
each way</li>
                   </ul>
            </li>

            
        </ul>

# <span style="color:green">Bandwidth – Gravity of Modern Computer Systems</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <body>
            
                <ul>
                    <li>Goal is to maintain DC
balance while have sufficient
state transition for clock
recovery</li>
                    <li>The difference of 1s and 0s in
a 20-bit stream should be ≤ 2</li>
                    <li>There should be no more
than 5 consecutive 1s or 0s in
any stream</li>
                    <li>00000000, 00000111, 11000001
bad</li>
                    <li>01010101, 11001100 good</li>
                    <li>Find 256 good patterns among
1024 total patterns of 10 bits to
encode an 8-bit data</li>
                    <li>20% overhead</li>
                </ul>
       </body>

</div>

# <span style="color:green">PCIe Data Transfer using DMA</span>


<br></br>
<div style="text-align:justify;font-size:23px">

        
        
        <ul>
            <li>DMA (Direct Memory
Access) is used to fully
utilize the bandwidth of an
I/O bus
                   <ul>
                       <li>DMA uses physical address for
source and destination
                        </li>
                       <li>Transfers a number of bytes
requested by OS</li>
                        <li>Needs pinned memory</li>
                        <li>DMA hardware is much faster than
CPU software and frees the CPU for
other tasks during the data transfer</li>
                   </ul>
            </li>

            
        </ul>

# <span style="color:green">Pinned Memory</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <body>
            
                <ul>
                    <li>DMA uses physical addresses</li>
                    <li>The OS could accidentally page
out the data that is being read or
written by a DMA and page in
another virtual page into the
same location</li>
                    <li>Pinned memory cannot not be
paged out</li>
                    <li>If a source or destination of a
cudaMemCpy() in the host
memory is not pinned, it needs
to be first copied to a pinned
memory – extra overhead</li>
                    <li>cudaMemcpy is much faster
with pinned host memory
source or destination</li>
                </ul>
       </body>

</div>

# <span style="color:green">Allocate/Free Pinned Memory (a.k.a. Page Locked Memory)</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <ul>
            <body>
            <li>cudaHostAlloc()
                <ul>
                    <li>Three parameters</li>
                    <li>Address of pointer to the allocated memory</li>
                    <li>Size of the allocated memory in bytes</li>
                    <li>Option – use cudaHostAllocDefault for now</li>
                </ul>
            </li>
            <li>cudaFreeHost()
                <ul>
                    <li>One parameter</li>
                    <li>Pointer to the memory to be freed</li>
                </ul>
            </li>
          
            </body>
        </ul>
</div>

# <span style="color:green">Using Pinned Memory</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <body>
            
                <ul>
                    <li>Use the allocated memory and its pointer the same way those
returned by malloc();</li>
                    <li>The only difference is that the allocated memory cannot be paged by
the OS</li>
                    <li>The cudaMemcpy function should be about 2X faster with pinned
memory</li>
                    <li>Pinned memory is a limited resource whose over-subscription can
have serious consequences</li>
                </ul>
       </body>
