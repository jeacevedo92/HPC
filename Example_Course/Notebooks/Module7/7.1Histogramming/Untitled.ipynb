# <span style="color:green">Histogram</span>

<br></br>
<div style="text-align:justify">
        <ul>
            <h2>
            <li>A method for extracting notable features and patterns from large data sets
                <ul>
                    <li>Feature extraction for object recognition in images</li>
                    <li>Fraud detection in credit card transactions</li>
                    <li>Correlating heavenly object movements in astrophysics</li>
                    <li>...</li>
                </ul>
            </li>
            <li>Basic histograms - for each element in the data set, use the value to
identify a “bin counter” to increment</li>
            </h2>
        </ul>
</div>

# <span style="color:green">A simple parallel histogram algorithm</span>


<br></br>
<div style="text-align:justify">
        <ul>
            <h2>
                    <li>Partition the input into sections</li>
                    <li>Have each thread to take a section of the input</li>
                    <li>Each thread iterates through its section.</li>
                    <li>For each letter, increment the appropriate bin counter</li>
            </h2>
        </ul>
</div>

# <span style="color:green">Input Partitioning Affects Memory Access Efficiency</span>


<br></br>
<div style="text-align:justify">

        
        
        <ul>
            <h2>
            <li>Sectioned partitioning results in poor memory access efficiency
                <ul>
                    <li>Adjacent threads do not access adjacent memory locations</li>
                    <li>Accesses are not coalesced</li>
                    <li>DRAM bandwidth is poorly utilized</li>
                   
                    
                </ul>
            </li>
            </h2>
        </ul>

            <h2>
            <li>Change to interleaved partitioning
                <ul>
                    <li>All threads process a contiguous section of elements</li>
                    <li>They all move to the next section and repeat</li>
                    <li>The memory accesses are coalesced</li>
