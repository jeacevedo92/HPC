# <span style="color:green">Cost and Benefit of Privatization</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <ul>
            <body>
            
            <li>Cost
                <ul>
                    <li>Overhead for creating and initializing private copies</li>
                    <li>Overhead for accumulating the contents of private copies into the final copy</li>
                </ul>
            </li>
            <li>Benefit
                <ul>
                    <li>Much less contention and serialization in accessing both the private copies and the
final copy</li>
                    <li>The overall performance can often be improved more than 10x</li>
                </ul>
            </li>
            </body>
        </ul>
</div>

<hr style="height:2px">

# <span style="color:green">Shared Memory Atomics for Histogram</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <ul>
            <body>
            
            <li>Each subset of threads are in the same block</li>
            <li>Much higher throughput than DRAM (100x) or L2 (10x) atomics</li>
            <li>Less contention â€“ only threads in the same block can access a
shared memory variable</li>
            <li>This is a very important use case for shared memory!</li>
            </body>
        </ul>
</div>

# <span style="color:green">More on Privatization</span>

<br></br>
<div style="text-align:justify;font-size:20px">
        <ul>
            <body>
            <li>Privatization is a powerful and frequently used technique for
parallelizing applications</li>
            <br></br>
            <li>The operation needs to be associative and commutative
                <ul>
                    <li>Histogram add operation is associative and commutative</li>
                    <li>No privatization if the operation does not fit the requirement</li>
                </ul>
            </li>
            <br></br>
            <li>The private histogram size needs to be small
                <ul>
                    <li>Fits into shared memory</li>
                </ul>
            </li>
            <br></br>
            <li>What if the histogram is too large to privatize?
                <ul>
                    <li>Sometimes one can partially privatize an output histogram and use range
testing to go to either global memory or shared memory</li>
                </ul>
            </li>
            </body>
        </ul>
</div>
